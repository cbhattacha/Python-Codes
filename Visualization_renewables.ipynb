```python
import pandas as pd

df_solar=pd.read_csv('solar-15m_2019-09-01_2020-09-30.csv',parse_dates=True)
df_solar['time']=pd.to_datetime(df_solar.time,dayfirst=True)
df_solar.set_index('time',inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

```

```python
plt.rcParams['figure.figsize'] = (10, 8)   # Increases the Plot Size
df_solar.plot(grid = True)
```

```python
df_solar_mean=df_solar.resample(rule='M').mean()
plt.rcParams['figure.figsize'] = (10, 8)   # Increases the Plot Size
df_solar_mean.plot(grid = True)
```

```python
plt.rcParams['figure.figsize'] = (10, 8) # change plot size
df_solar.resample('M').max().plot(kind='bar')
plt.title('Monthly Maximum Values')
```

```python
#Time series lag plot
pd.plotting.lag_plot(df_solar['Solarenergie_Hochrechnung_50Hertz'])
```

```python
df_wind=pd.read_csv('wind-15m_2019-09-01_2020-09-30.csv',parse_dates=True)
df_wind['time']=pd.to_datetime(df_wind.time,dayfirst=True)
df_wind.set_index('time',inplace=True)
```

```python
#Time series lag plot
pd.plotting.lag_plot(df_wind['Windenergie_Hochrechnung_TransnetBW'])
```
